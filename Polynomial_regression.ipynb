```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

# Sample data (1 feature, 1 target)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1.2, 1.9, 3.2, 4.1, 5.0])

# Step 1: Create polynomial features
degree = 3  # M-degree (polynomial)
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

# Step 2: Apply Ridge Regression (L2 regularization)
ridge_reg = Ridge(alpha=0.1)  # alpha = λ (regularization strength)
ridge_reg.fit(X_poly, y)

# Predictions
y_pred = ridge_reg.predict(X_poly)

# Output results
print("Predicted values:", y_pred)
print("Mean Squared Error:", mean_squared_error(y, y_pred))
print("Ridge Regression Coefficients:", ridge_reg.coef_)

# Plot the results
plt.scatter(X, y, color='blue', label="Original data")  # Original data points
plt.plot(X, y_pred, color='red', label="Ridge Regression fit")  # Predicted curve

# Add labels and title
plt.title('Polynomial Regression with L2 Regularization (Ridge)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
